```python
import pandas as pd
import matplotlib.pyplot as plt
import functools as ft
import numpy as np
from wordcloud import WordCloud, STOPWORDS
from collections import Counter
import re
import nltk
```

```python
severe_combine=  pd.read_csv('../data/severe_new.csv')
```

```python
severe_combine.columns
```

```python
female = severe_combine.loc[severe_combine['SEX']=='F']
```

```python
female
```

```python
a = female.SYMPTOM1.value_counts().to_frame()
```

```python
a = a.loc[a['SYMPTOM1']>1000]
```

```python
a.reset_index(inplace = True)
```

```python
b = female.SYMPTOM2.value_counts().to_frame()
```

```python
b= b.loc[b['SYMPTOM2'] >1000]
```

```python
b.reset_index(inplace = True)
```

```python
c = female.SYMPTOM3.value_counts().to_frame()
c = c.loc[c['SYMPTOM3']>1000]
```

```python
c.reset_index(inplace = True)
```

```python
dfs = [a, b, c]
symptoms = ft.reduce(lambda left, right: pd.merge(left, right, on='index'), dfs)
```

```python
symptoms
```

```python
symptoms.to_csv('../data/symptomsf_tab.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import functools as ft
```

```python
v_data_2019 = pd.read_csv('2019VAERSData.csv')
```

```python
v_vax_2019 = pd.read_csv('2019VAERSVAX.csv')
```

```python
v_symptoms_2019 = pd.read_csv('2019VAERSSYMPTOMS.csv')
```

```python
dfs = [v_data_2019, v_vax_2019,v_symptoms_2019]
vaccination_2019 = ft.reduce(lambda left, right: pd.merge(left, right, on='VAERS_ID'), dfs)
```

```python
vaccination_2019.head()
```

```python
vaccination_2019.columns
```
